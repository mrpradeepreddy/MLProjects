````
# %% [markdown]
# # Real-Time Driver Drowsiness & Emotion Detection System
# **Tech Stack:** MediaPipe, OpenCV, TensorFlow (CNN+LSTM), NumPy
#
# ## Project Overview
# This notebook implements a safety system that monitors a driver's face to detect:
# 1. **Drowsiness:** via Eye Aspect Ratio (EAR), Mouth Aspect Ratio (MAR), and Head Tilt.
# 2. **Emotion:** via a Deep Learning model (CNN + LSTM) to detect negative emotional states (Anger, Sadness, Fear) that might affect driving.
#
# ---

# %% [markdown]
# ## 1. Setup & Installation
# We install the necessary libraries for computer vision and facial landmark detection.

# %%
# !pip install mediapipe opencv-python tensorflow numpy matplotlib

import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, LSTM, TimeDistributed, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from scipy.spatial import distance as dist
import os
from google.colab import files  # Specific to Colab for file handling

print(f"TensorFlow Version: {tf.__version__}")
print(f"MediaPipe Version: {mp.__version__}")

# Create necessary directories
os.makedirs('models', exist_ok=True)
os.makedirs('output', exist_ok=True)

# %% [markdown]
# ## 2. Geometric Logic (EAR, MAR, Head Tilt)
# We define the mathematical formulas to calculate aspect ratios from facial landmarks.

# %%
def calculate_ear(eye_points):
    """
    Calculates Eye Aspect Ratio (EAR).
    EAR = (|p2-p6| + |p3-p5|) / (2 * |p1-p4|)
    """
    # Vertical distances
    A = dist.euclidean(eye_points[1], eye_points[5])
    B = dist.euclidean(eye_points[2], eye_points[4])
    # Horizontal distance
    C = dist.euclidean(eye_points[0], eye_points[3])

    ear = (A + B) / (2.0 * C)
    return ear

def calculate_mar(mouth_points):
    """
    Calculates Mouth Aspect Ratio (MAR).
    MAR = (|p2-p8| + |p3-p7| + |p4-p6|) / (2 * |p1-p5|)
    Note: Using simplified height/width for MediaPipe specific points
    """
    # Vertical distance (Top lip to bottom lip)
    A = dist.euclidean(mouth_points[1], mouth_points[3])
    # Horizontal distance (Left corner to right corner)
    B = dist.euclidean(mouth_points[0], mouth_points[2])

    mar = A / B
    return mar

def get_head_tilt(nose_tip, chin_tip):
    """
    Simple geometric estimation of head tilt using x-coordinates of nose and chin.
    If aligned vertically, x diff is small.
    """
    tilt_angle = np.arctan2(nose_tip[0] - chin_tip[0], nose_tip[1] - chin_tip[1])
    return np.degrees(tilt_angle)

# MediaPipe Face Mesh Indices (Standard 468 points)
# Left Eye: [33, 160, 158, 133, 153, 144] (P1..P6)
LEFT_EYE_IDX = [33, 160, 158, 133, 153, 144]
# Right Eye: [362, 385, 387, 263, 373, 380]
RIGHT_EYE_IDX = [362, 385, 387, 263, 373, 380]
# Mouth (Outer): [61, 13, 291, 14] -> [Left, Top, Right, Bottom]
MOUTH_IDX = [61, 13, 291, 14]
NOSE_TIP_IDX = 1
CHIN_IDX = 152

# %% [markdown]
# ## 3. Emotion Recognition Model (CNN + LSTM)
# We build a model that processes a *sequence* of frames. The CNN extracts features from each frame, and the LSTM aggregates them over time to predict emotion.
#
# **Note on Data:** # In a real scenario, you would mount Google Drive and load the FER-2013 dataset.
# For this "Runnable" notebook, we generate **Synthetic Data** so you can see the pipeline run immediately without downloading 100MB+ data.

# %%
# Configuration
IMG_SIZE = 48
SEQ_LENGTH = 5  # Look at last 5 frames for emotion context
NUM_CLASSES = 7 # Happy, Neutral, Sad, Anger, Fear, Surprise, Disgust
CLASSES = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

def build_cnn_lstm_model():
    model = Sequential()

    # --- CNN Layers (Wrapped in TimeDistributed) ---
    # This allows the CNN to process every frame in the sequence independently
    model.add(TimeDistributed(Conv2D(32, (3, 3), activation='relu'),
                              input_shape=(SEQ_LENGTH, IMG_SIZE, IMG_SIZE, 1)))
    model.add(TimeDistributed(BatchNormalization()))
    model.add(TimeDistributed(MaxPooling2D(pool_size=(2, 2))))

    model.add(TimeDistributed(Conv2D(64, (3, 3), activation='relu')))
    model.add(TimeDistributed(BatchNormalization()))
    model.add(TimeDistributed(MaxPooling2D(pool_size=(2, 2))))

    model.add(TimeDistributed(Flatten()))

    # --- LSTM Layers ---
    # This processes the sequence of features from the CNN
    model.add(LSTM(64, return_sequences=False))
    model.add(Dropout(0.5))

    # --- Output ---
    model.add(Dense(NUM_CLASSES, activation='softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

model = build_cnn_lstm_model()
model.summary()

# %% [markdown]
# ### 3.1 Train the Model (Synthetic/Demo Mode)
# We create random data to demonstrate the training loop works.
# *To use real data: Replace `X_train` with loaded FER-2013 images reshaped to (N, 5, 48, 48, 1).*

# %%
print("Generating synthetic training data for demonstration...")
# 100 samples, sequence length 5, 48x48 grayscale images
X_train_dummy = np.random.rand(100, SEQ_LENGTH, IMG_SIZE, IMG_SIZE, 1).astype('float32')
y_train_dummy = to_categorical(np.random.randint(0, NUM_CLASSES, 100), num_classes=NUM_CLASSES)

print("Training Model (Demo Run)...")
history = model.fit(
    X_train_dummy, y_train_dummy,
    epochs=5,  # Short run for demo
    batch_size=10,
    validation_split=0.2,
    verbose=1
)

# Save the model
model.save('models/cnn_lstm_emotion.h5')
print("Model saved to models/cnn_lstm_emotion.h5")

# Plot Training Results
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# %% [markdown]
# ## 4. Video Inference Pipeline
# This is the core system. It processes an uploaded video frame-by-frame.

# %%
# --- Config Thresholds ---
EAR_THRESHOLD = 0.25    # Eyes closed
MAR_THRESHOLD = 0.5     # Yawning (Wide open mouth)
CLOSED_EYES_FRAME_LIMIT = 15  # ~0.5 to 1 second depending on FPS
TILT_THRESHOLD = 15     # Degrees

# --- Alert System Mockup ---
def trigger_alert(message, method="print"):
    """
    Simulate an alert. In production, integrate Twilio/Firebase here.
    """
    if method == "print":
        print(f"⚠️ ALERT: {message}")
    elif method == "twilio":
        # Twilio Code Placeholder
        # client.messages.create(body=message, from_='+1234', to='+5678')
        pass

def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Video Writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # MediaPipe Setup
    mp_face_mesh = mp.solutions.face_mesh
    face_mesh = mp_face_mesh.FaceMesh(
        refine_landmarks=True,
        max_num_faces=1,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
    )

    # State Variables
    frame_count = 0
    blink_counter = 0
    drowsy_frames = 0
    yawn_frames = 0

    # Emotion Buffer (Rolling window of frames)
    emotion_frame_buffer = []
    current_emotion = "Neutral"

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_mesh.process(frame_rgb)

        h, w, _ = frame.shape

        # Default Status
        status = "Alert"
        color = (0, 255, 0) # Green

        if results.multi_face_landmarks:
            for face_landmarks in results.multi_face_landmarks:
                lm = face_landmarks.landmark

                # 1. Convert landmarks to numpy array (x, y)
                points = np.array([(int(p.x * w), int(p.y * h)) for p in lm])

                # 2. Calculate Drowsiness Metrics
                left_eye = points[LEFT_EYE_IDX]
                right_eye = points[RIGHT_EYE_IDX]
                mouth = points[MOUTH_IDX]
                nose = points[NOSE_TIP_IDX]
                chin = points[CHIN_IDX]

                ear_left = calculate_ear(left_eye)
                ear_right = calculate_ear(right_eye)
                ear_avg = (ear_left + ear_right) / 2.0

                mar = calculate_mar(mouth)

                tilt = get_head_tilt(nose, chin)

                # 3. Drowsiness Logic
                # Check Eyes
                if ear_avg < EAR_THRESHOLD:
                    blink_counter += 1
                    if blink_counter > CLOSED_EYES_FRAME_LIMIT:
                        status = "DROWSY (Eyes Closed)"
                        color = (0, 0, 255) # Red
                        trigger_alert("Driver Eyes Closed!", method="print")
                else:
                    blink_counter = 0

                # Check Yawn
                if mar > MAR_THRESHOLD:
                    yawn_frames += 1
                    if yawn_frames > 10:
                        status = "DROWSY (Yawning)"
                        color = (0, 0, 255)
                else:
                    yawn_frames = 0

                # Check Head Nod (Tilt)
                if abs(tilt) > TILT_THRESHOLD:
                    status = "Distracted/Nodding"
                    color = (0, 165, 255) # Orange

                # 4. Emotion Detection Logic
                # Preprocess face for model
                # Extract bounding box of face loosely based on landmarks
                x_min, y_min = np.min(points, axis=0)
                x_max, y_max = np.max(points, axis=0)

                # Ensure bbox is valid
                x_min, y_min = max(0, x_min), max(0, y_min)
                x_max, y_max = min(w, x_max), min(h, y_max)

                face_roi = cv2.cvtColor(frame[y_min:y_max, x_min:x_max], cv2.COLOR_BGR2GRAY)

                try:
                    face_roi = cv2.resize(face_roi, (IMG_SIZE, IMG_SIZE))
                    face_roi = face_roi / 255.0
                    face_roi = np.expand_dims(face_roi, axis=-1) # (48, 48, 1)

                    emotion_frame_buffer.append(face_roi)

                    # Keep buffer at SEQ_LENGTH
                    if len(emotion_frame_buffer) > SEQ_LENGTH:
                        emotion_frame_buffer.pop(0)

                    # Predict every 5 frames if buffer is full
                    if len(emotion_frame_buffer) == SEQ_LENGTH and frame_count % 5 == 0:
                        input_seq = np.array([emotion_frame_buffer]) # (1, 5, 48, 48, 1)
                        preds = model.predict(input_seq, verbose=0)
                        current_emotion = CLASSES[np.argmax(preds)]
                except Exception as e:
                    pass # ROI resize issues possible if face lost

                # 5. Drawing on Frame
                cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), color, 2)

                # Draw Eye Contours
                cv2.polylines(frame, [left_eye], True, (255, 255, 255), 1)
                cv2.polylines(frame, [right_eye], True, (255, 255, 255), 1)
                cv2.polylines(frame, [mouth], True, (255, 255, 255), 1)

                # UI Text
                cv2.putText(frame, f"Status: {status}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
                cv2.putText(frame, f"Emotion: {current_emotion}", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
                cv2.putText(frame, f"EAR: {ear_avg:.2f} | MAR: {mar:.2f}", (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

        out.write(frame)

    cap.release()
    out.release()
    print(f"Processing complete. Video saved to: {output_path}")

# %% [markdown]
# ## 5. Run with Custom Video
# Execute this cell to upload a video file and run the detection system.

# %%
# Upload Video
print("Please upload a video file (mp4) of a face...")
uploaded = files.upload()
video_name = list(uploaded.keys())[0]

output_filename = 'output_annotated.mp4'

# Run Pipeline
process_video(video_name, output_filename)

# Download Result
files.download(output_filename)

# %% [markdown]
# ## 6. Project Structure & Documentation
#
# ### Project Tree
# ```
# /driver_drowsiness_colab/
# │── driver_drowsiness_colab.py    # Main script
# │── models/
# │   └── cnn_lstm_emotion.h5       # Trained Emotion Model
# │── output/
# │   └── output_annotated.mp4      # Result video
# │── dataset/                      # (Optional) Place FER-2013 here
# ```
#
# ### Summary
# This system uses a hybrid approach:
# 1.  **Deterministic CV (MediaPipe):** Extremely fast and accurate for calculating geometric features (EAR/MAR) to detect physical signs of sleepiness immediately.
# 2.  **Probabilistic DL (CNN-LSTM):** Analyzes temporal context to understand emotional states that may lead to aggressive or distracted driving.
#
# ### Limitations & Future Scope
# * **Lighting:** MediaPipe is robust, but extreme darkness requires IR cameras.
# * **Optimization:** The CNN-LSTM model in this notebook uses a small architecture. For production, use MobileNetV2 for feature extraction.
# * **Real-Time:** To run this on a local machine (not Colab), simply replace the `files.upload()` block with `cap = cv2.VideoCapture(0)` to use the webcam.
````
